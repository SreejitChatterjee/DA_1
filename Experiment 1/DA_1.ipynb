```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(10)
plt.plot(df['COUNTRY'][0:20],df['PERSONS_FULLY_VACCINATED_PER100'][0:20])
plt.show()
```

```python
sns.pairplot(df[0:20],vars = ['PERSONS_FULLY_VACCINATED_PER100','PERSONS_VACCINATED_1PLUS_DOSE' ])
```

```python
sns.relplot(x = 'TOTAL_VACCINATIONS', y = 'TOTAL_VACCINATIONS_PER100',hue = 'NUMBER_VACCINES_TYPES_USED',data = df[0:50])
```

```python
sns.relplot(y = 'NUMBER_VACCINES_TYPES_USED', x = 'PERSONS_FULLY_VACCINATED_PER100',hue = 'WHO_REGION',data = df[0:50])
```

```python
sns.distplot(df['PERSONS_FULLY_VACCINATED_PER100'][0:200],bins = 30)
```

```python
sns.distplot(df['NUMBER_VACCINES_TYPES_USED'][0:200],bins = 20)
```

#INFERENCE

From this dataset we first get a rough idea as to how many null values there are and we eliminate them along with unrequired columns.

Then we go on to plot a line graph in which we see a trend that African countries tend to have lower vaccination rates than others.

After this using sns.relplot() we can infer that EMRO WHO_REGION varies the different vaccine types it uses, the most along with SEARO, while AFRO WHO_REGIONS mostly use a less varied vaccine range. Also from this plot we can get a better idea as to how the distribution of vaccines have gone throughout the world.

In this dataset by using a distplot we see that the highest density of people fully vaccinated per 100 is in a range of 70-80 which shows that the vaccination of the world is going at a satisfactory pace.

Lastly we see that most countries have opted to use 4 different types of vaccines for their citizens
